```python
# 初めに1回だけ実行する
# これをしないと以降の操作ができない

# 日本語の文章を単語に分けるためのツール（Tokenizer）をインポートする
from janome.tokenizer import Tokenizer

# ワードクラウドを作るためのツール（WordCloud）をインポートする
from wordcloud import WordCloud

# グラフを表示するためのツール（matplotlib.pyplot）をインポートする
import matplotlib.pyplot as plt
```

```python
# 初めに1回だけ実行する
# ワードクラウドを作成する関数（機能）を作成する
# これがないと以降の操作ができない

def make_wordcloud(text):
    # textを単語ごとに区切って（形態素解析）、その結果を「tokens」に入れる
    t = Tokenizer()
    tokens = t.tokenize(text)

    # 単語を格納するための空のリストを2つ作る
    words_all = []
    words = []

    # 区切られた単語（トークン）を一つずつ取り出して処理する
    for token in tokens:
        # 改行文字 (\n) 以外の単語の表層形（実際に文章に書かれている形）をリストに追加する
        if token.surface != "\n":
            words_all.append([token.surface, token.part_of_speech])

        # 品詞情報（例：「名詞,一般,*,*」）を「,」で区切って、最初の要素（例：名詞）と2番目の要素（例：一般）を取り出す
        hinshi = token.part_of_speech.split(",")[0]
        hinshi2 = token.part_of_speech.split(",")[1]

        if hinshi == "名詞":
            if hinshi2 != "数" and hinshi2 != "代名詞" and hinshi2 != "非自立":
                # その単語の表層形をリストに追加する
                words.append(token.surface)

        if hinshi == "動詞" or hinshi == "形容詞" or hinshi == "副詞":
            # その単語の原形をリストに追加する
            words.append(token.base_form)

    print(words_all)

    # リストに入っている単語を半角スペース区切りでつなげて一つの文字列にする（分かち書き）
    words_wakachi = " ".join(words)

    # ワードクラウドに表示したくない単語（ストップワード）のリストを定義する
    # ★★★★適宜追加してください★★★★
    stop_words = ["する", "れる", "いる", "居る", "くれる", "なる", "ある", "在る", "ない", "無い", "やる", "される", "やすい", "そう", "いける", "きれる", "なくなる", "られる", "つて", "ふる", "ゐる", "できる", "さ", "にくい", "すぎる", "思う", "おもう", "もらう", "いい", "みんな", "この", "やばい", "話", "知る", "てる", "これ", "は", "今", "から", "50", "年", "以上", "前", "の", "出来事", "です" ]

    # フォントなどの設定をして、上の文字列「words_wakachi」を使ってワードクラウドを生成する
    word_cloud = WordCloud(stopwords=stop_words, font_path="/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc", width=1250, height=490, min_font_size=5, collocations=False, background_color="#ffffff", max_words=400).generate(words_wakachi)

    # 表示する画像のサイズを決める
    figure = plt.figure(figsize=(15, 10))

    # ワードクラウドの設定を行う
    plt.imshow(word_cloud)
    plt.tick_params(labelbottom=False, labelleft=False)
    plt.xticks([])
    plt.yticks([])

    # 設定したワードクラウドを画面に表示する
    plt.show()
```

## ワードクラウド作成

「事前準備」のコードを全て実行してから、以下のコードを実行してください。テキスト入力欄が表示されるので、ワードクラウドにしたい文章を入力またはコピー＆ペースト（`Ctrl` + `C` → `Ctrl` + `V`）してください。スプレッドシートからのペーストには `Ctrl` + `Shift` + `V` を使ってください。

```python
# input関数を使って、ワードクラウドにしたい文章を入力する
# スプレッドシートからのペーストは【 Ctrl + Shift + V 】
text = input("ワードクラウドにしたい文章を入力してエンターキーを押してください。")
make_wordcloud(text)
```
